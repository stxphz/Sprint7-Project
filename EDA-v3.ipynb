<h3><u>Project Description</u></h3>

Sprint 7's project revolves around a collection of datasets sourced from Instacart, a prominent grocery delivery platform akin to services like Uber Eats and DoorDash. These datasets were initially made publicly available by Instacart in 2017 for a Kaggle competition.

This version of the datasets have been thoughtfully adapted by Practicum. In order to enhance computational efficiency, the dataset size was reduced. Practicum also deliberately introduced missing and duplicate values while meticulously preserving the original data's distribution characteristics.

The provided datasets are named: <code>aisles.csv</code>, <code>departments.csv</code>, <code>instacart_orders.csv</code>, <code>order_products.csv</code>, and <code>products.csv</code>.

The primary objective of this project is to firstly, perform comprehensive data cleaning. Secondly, to compile a report that offers valuable insights into the shopping behaviors exhibited by Instacart customers. This project also requires the creation of informative visualizations to effectively convey the discoveries derived from the provided datasets. 

<br></br>

<h3><u>Data Dictionary</u></h3>

<code>aisles.csv</code>:
- <code>'aisle_id'</code>: ID number that uniquely identifies each grocery aisle category
- <code>'aisle'</code>: name of the aisle

<code>departments.csv</code>:
- <code>'department_id'</code>: ID number that uniquely identifies each grocery department category
- <code>'department'</code>: name of the department

<code>instacart_orders.csv</code>: each row corresponds to one order on the Instacart app
- <code>'order_id'</code>: ID number that uniquely identifies each order
- <code>'user_id'</code>: ID number that uniquely identifies each customer account
- <code>'order_number'</code>: the number of times this customer has placed an order
- <code>'order_dow'</code>: day of the week that the order placed (which day is 0 is uncertain)
- <code>'order_hour_of_day'</code>: hour of the day that the order was placed
- <code>'days_since_prior_order'</code>: number of days since this customer placed their previous order

<code>order_products.csv</code>: each row corresponds to one item placed in an order
- <code>'order_id'</code>: ID number that uniquely identifies each order
- <code>'product_id'</code>: ID number that uniquely identifies each product
- <code>'add_to_cart_order'</code>: the sequential order in which each item was placed in the cart
- <code>'reordered'</code>: 0 if the customer has never ordered this product before, 1 if they have

<code>products.csv</code>: each row corresponds to a unique product that customers can buy
- <code>'product_id'</code>: ID number that uniquely identifies each product
- <code>'product_name'</code>: name of the product
- <code>'aisle_id'</code>: ID number that uniquely identifies each grocery aisle category
- <code>'department_id'</code>: ID number that uniquely identifies each grocery department category

```python
import pandas as pd
import plotly_express as px

# filter rows where 'days_since_prior_order is not null 
# (this is to exclude the customer's first orders)
filtered_orders = df_instacart_orders[df_instacart_orders['days_since_prior_order'].notnull()]

# create a histogram plot  using Plotly Express
fig = px.histogram(filtered_orders,
                   x='days_since_prior_order',
                   nbins=50
                   )

fig.update_layout(title='Distribution of Days Since Prior Order',
                  xaxis_title='Days Since Prior Order',
                  yaxis_title='Frequency'
                  )

# comment the minimum and maximum values with annotations
min_days = filtered_orders['days_since_prior_order'].min()
max_days = filtered_orders['days_since_prior_order'].max()
fig.add_annotation(text=f'Minimum:{min_days} days', x=min_days, y=100, showarrow=True, arrowhead=1, arrowcolor="red", font=dict(color="red"))
fig.add_annotation(text=f'Maximum:{max_days} days', x=max_days, y=100, showarrow=True, arrowhead=1, arrowcolor="red", font=dict(color="red"))

fig.show()
```

<h1>[B]1.</h1>

> Is there a difference in <code>'order_hour_of_day'</code> distributions on Wednesdays and Saturdays? Plot the histograms for both days on the same plot and describe the differences that you see.

```python
import pandas as pd
import plotly_express as px

# df_instacart_orders
# 'order_hour_of_day == 'order_placed_hour_of_day'
# 'order_dow' == 'order_placed_day_of_week
# 0=Monday , 1=Tuesday , 2=Wednesday , 3=Thursday , 4=Friday , 5=Saturday , 6=Sunday (assuming this dataset's convention follows the ISO weekday numbering)

# filter data for Wednesday and Saturday
wednesday_data = df_instacart_orders[df_instacart_orders['order_placed_day_of_week'] == 2]
saturday_data = df_instacart_orders[df_instacart_orders['order_placed_day_of_week'] == 5]

fig = px.histogram(
    wednesday_data, x='order_placed_hour_of_day',
    opacity=0.6,
    title='Distribution of Order Hour of Day (Wednesday vs. Saturday)',
    labels={'order_placed_hour_of_day': 'Hour of the Day'}
)

fig2 = px.histogram(
    saturday_data, x='order_placed_hour_of_day',
    opacity=0.6,
   labels={'order_placed_hour_of_day': 'Hour of the Day'}
)

fig.update_layout(xaxis_title = 'Hour of the Day',
                  yaxis_title = 'Frequency',
                  showlegend=True,
                  legend_title="Legend",                
)

# set names for the legend
fig.update_traces(name='Wednesday', showlegend=True)
fig2.update_traces(name='Saturday', showlegend=True, marker_color='red')

# combine the two traces in one plot
fig.add_traces(fig2.data)

fig.show()
```

> - although Saturday has a higher volume in frequency compared to Monday, they both have very similar distribution of peak times and run off times based on the overall shape.

> What are the top 20 items that are reordered most frequently (display their names and product IDs)?

```python
import pandas as pd

# df_order_products
# 'reordered' == 'product_reordered"

# group by 'product_id' and calculate the sum of 'reordered' to get the total reorders for each product
reorder_counts = df_order_products.groupby('product_id')['product_reordered'].sum().reset_index()

# merge reorder counts with the product names
top_reordered_products = pd.merge(reorder_counts, df_products[['product_id', 'product_name']], on='product_id')

# sort products by reorder count in descending order and select the top 20 reordered products
top_20_reordered_products = top_reordered_products.sort_values(by='product_reordered', ascending=False).head(20)

# display top 20 reordered products with their names and IDs
print(top_20_reordered_products[['product_id', 'product_name']])
```

<h1>[C]3.</h1>

> For each product, what proportion of its orders are reorders (create a table with columns for the product ID, product name, and reorder proportion)?

```python
import pandas as pd

# df_order_products
# 'reordered' == 'product_reordered'
# df_products

# calculate total number of orders for each product
total_orders_per_product = df_order_products.groupby('product_id')['product_reordered'].count().reset_index()

# calculate total number of reorder orders for each product
reorder_orders_per_product = df_order_products.groupby('product_id')['product_reordered'].sum().reset_index()

# merge the two dataFrames to get the total and reorder counts for each product
product_orders = pd.merge(total_orders_per_product, reorder_orders_per_product, on='product_id', suffixes=('_total', '_reorder'))

# calculate the reorder proportion for each product
product_orders['reorder_proportion'] = product_orders['product_reordered_reorder'] / product_orders['product_reordered_total']

# merge with the 'df_products' DataFrame to get product names
product_orders_with_names = pd.merge(product_orders, df_products[['product_id', 'product_name']], on='product_id')

# create a table with columns for product ID, product name, and reorder proportion
reorder_proportion_table = product_orders_with_names[['product_id', 'product_name', 'reorder_proportion']]

display(reorder_proportion_table)
```

<h1>[C]4.</h1>

> For each customer, what proportion of their products ordered are reorders?

```python
import pandas as pd

# df_orders_products
# 'reordered' == 'product_reordered'

# df_instacart_orders
# 'order_hour_of_day == 'order_placed_hour_of_day'
# 'order_dow' == 'order_placed_day_of_week

# merge 'df_order_products' and 'df_instacart_orders' to associate each product with a customer
merged_df = pd.merge(df_order_products, df_instacart_orders, on='order_id')

# calculate the total number of products ordered by each customer
total_products_per_customer = merged_df.groupby('user_id')['product_id'].count().reset_index()
total_products_per_customer.columns = ['user_id', 'total_products_ordered']

# calculate the total number of reorder products for each customer
reorder_products_per_customer = merged_df.groupby('user_id')['product_reordered'].sum().reset_index()
reorder_products_per_customer.columns = ['user_id', 'total_reorder_products']

# merge the two DataFrames to get the total and reorder counts for each customer
customer_orders = pd.merge(total_products_per_customer, reorder_products_per_customer, on='user_id')

# calculate the proportion of products ordered that are reorders for each customer
customer_orders['proportion_reorders'] = customer_orders['total_reorder_products'] / customer_orders['total_products_ordered']

# display customer_orders dataFrame with user_id and proportion_reorders columns
display(customer_orders[['user_id', 'proportion_reorders']])
```

<h1>[C]5.</h1>

> What are the top 20 items that people put in their carts first (display the product IDs, product names, and number of times they were the first item added to the cart)?

```python
import pandas as pd

# df_orders_products
# 'reordered' == 'product_reordered'
# df_products
# 'product_name' == 'product_name_lowercase'

# filter the df_order_products to include only products added first to the cart
first_in_cart = df_order_products[df_order_products['add_to_cart_order'] == 1]

# count the occurrences of each product
product_counts = first_in_cart['product_id'].value_counts().reset_index()
product_counts.columns = ['product_id', 'count']

# merge with 'df_products' to get product names
top_products_first_in_cart = pd.merge(product_counts, df_products, on='product_id')

# sort and select the top 20 products
top_20_first_in_cart = top_products_first_in_cart.sort_values(by='count', ascending=False).head(20)

# display table with product_id, product_name, and count of each product
display(top_20_first_in_cart[['product_id', 'product_name', 'count']])
```
