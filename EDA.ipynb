```python
# read instacart_orders.csv file and storing it to a variable 'df_instacart_orders' for DataFrame

import pandas as pd

df_instacart_orders = pd.read_csv('instacart_orders.csv', sep=';', decimal='.')
display(df_instacart_orders)
```

```python
# obtain general information about data in instacart_orders DataFrame including the index dtype and columns, non-null values and memory usage.

df_instacart_orders.info()
```

> df_instacart_orders.info() NOTES:
> - order_id datatype is an integer
> - user_id datatype is an integer
> - *rename 'order_number' to 'number_of_orders_placed' to make it clearer*
> - *rename 'order_dow' to 'order_placed_day_of_week' to make it clearer*
> - *rename 'order_hour_of_day' to 'order_placed_hour_of_day' to make it clearer*
> - there are missing values in 'days_since_prior_order'
> - change days_since_prior_order datatypes to int64 


```python
# renaming columns:
import pandas as pd
df_instacart_orders = pd.read_csv('instacart_orders.csv', sep=';')


df_instacart_orders.rename(columns={'order_number': 'number_of_orders_placed',      # rename column 'order_number' to 'number_of_orders_placed'
                                    'order_dow': 'order_placed_day_of_week',        # rename column 'order_dow' to 'order_placed_day_of_week'
                                    'order_hour_of_day': 'order_placed_hour_of_day' # rename column 'order_hour_of_day' to 'order_placed_hour_of_day'
                                    }, inplace=True)

display(df_instacart_orders.head(10))
```

```python
# validate there are missing values in instacart_order dataset

print((df_instacart_orders).isna().sum())
```

```python
# counting missing values directly in 'days_since_priot_order'

print(df_instacart_orders['days_since_prior_order'].value_counts(dropna=False))
```

```python
# get information about the numeric columns in instacart_orders dataset:

print(df_instacart_orders.describe())
```

```python
# fill missing values in 'days_since_prior_order' by using its mean value:

mean_days_since_prior_order = df_instacart_orders['days_since_prior_order'].mean()
df_instacart_orders['days_since_prior_order'].fillna(mean_days_since_prior_order, inplace=True)
```

```python
# double check to see if missing values were successfully filled:

print((df_instacart_orders).isna().sum())
```

```python
# change datatype of 'days_since_prior_order' to int64

df_instacart_orders['days_since_prior_order'] = df_instacart_orders['days_since_prior_order'].astype('int')
print(df_instacart_orders.info())
```

```python
# display the table to look at instacart_orders dataset to see all changes made:

display(df_instacart_orders)
```
